```python
# replace with right names if you changed default file names in previous cell

input_txt_file = 'input.txt'
input_portrait = 'monalisa2.jpg'
```

```python
from IPython.display import HTML
from base64 import b64encode

output_video_name = '{0}_{1}.mp4'.format(
  input_portrait.split('.')[0],
  input_txt_file.split('.')[0]
  )
mp4 = open('{0}'.format(output_video_name),'rb').read()
data_url = "data:video/mp4;base64," + b64encode(mp4).decode()

display(HTML("""
<video width=600 controls>
      <source src="%s" type="video/mp4">
</video>
""" % data_url))
```
